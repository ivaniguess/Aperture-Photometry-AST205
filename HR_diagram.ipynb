```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("hr_data.csv")

plt.figure(figsize=(10,10))
plt.scatter(df["Color (g-r)"], df["M_abs (g)"], s=50, alpha=0.7, label="Sample")


ms_color = 0.391
ms_mag = 5.163

plt.scatter(ms_color, ms_mag, 
            s=120, 
            marker='*', 
            edgecolor='black',
            label="MS Star")

wd_color = -0.3859
wd_mag = 8.408

plt.scatter(wd_color, wd_mag, 
            s=120, 
            marker='*', 
            edgecolor='black',
            label="White Dwarf")

plt.xlabel("Color (g - r)")
plt.ylabel("Absolute Magnitude $M_g$")

plt.gca().invert_yaxis()

plt.legend()
plt.grid(alpha=0.3)

plt.tight_layout()
plt.show()
```
